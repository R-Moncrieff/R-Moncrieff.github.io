```python
import requests, json, csv, io
from requests.packages.urllib3.exceptions import InsecureRequestWarning

requests.packages.urllib3.disable_warnings(InsecureRequestWarning)

#function that gets the body of website and trucates to 3k charaters:
def get_website(website):
  #the required first parameter of the 'get' method is the 'url':
  response = requests.get(website, verify=False, allow_redirects=True)

  #find <body> in response
  response_text = response.text

  #print(response_text)
  body_start = response_text.find("<body")

  #get the text for GPT3
  body_text = response_text[body_start:(body_start + 3000)] + "  $$$$$$$$$$$"

  return(body_text)

#function that loads csv of websites and labels and runs the get_website function to create
training_dictorary = []
errors = []

with open('training_2.csv', mode='r') as csvfile:
  reader = csv.reader(csvfile)
  for row in reader:
    try:
      prompt = get_website(row[0])
      completion = row[1]
      training_dictorary.append({"prompt":prompt,"completion":completion})
    except:
      errors.append(row)

print(training_dictorary)
```

```python
number_of_errors = 0
for error in errors:
  print(error)
  number_of_errors += 1
```

```python
print(number_of_errors)
```

```python
keys = training_dictorary[0].keys()

with open('training_output.csv','w', newline='') as output_file:
  dict_writer = csv.DictWriter(output_file, keys)
  dict_writer.writeheader()
  dict_writer.writerows(training_dictorary)
```
